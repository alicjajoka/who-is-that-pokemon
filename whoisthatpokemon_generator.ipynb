```python
import cv2

bg = cv2.imread('bg.png', cv2.IMREAD_UNCHANGED)

for pokedex_number in range(1, 152):
    file_path = 'pokemon/'+str(pokedex_number)+'.png'
    img = cv2.imread(file_path, cv2.IMREAD_UNCHANGED)
    print(file_path)
    
    scale_percent = 250
    width = int(img.shape[1] * scale_percent / 100)
    height = int(img.shape[0] * scale_percent / 100)
    dsize = (width, height)
    img = cv2.resize(img, dsize)

    _, mask = cv2.threshold(img[:, :, 3], 0, 255, cv2.THRESH_BINARY_INV)
    mask = cv2.cvtColor(mask,cv2.COLOR_GRAY2RGBA)

    for i in range(len(mask)):
        for j in range(len(mask[i])):
            if all(mask[i][j] == [255,255,255,255]):
                mask[i][j][3]=0

    x_offset=y_offset=220
    y1, y2 = y_offset, y_offset + img.shape[0]
    x1, x2 = x_offset, x_offset + img.shape[1]
    alpha_s = img[:, :, 3] / 255.0
    alpha_l = 1.0 - alpha_s

    malpha_s = mask[:, :, 3] / 255.0
    malpha_l = 1.0 - malpha_s

    hidden=bg.copy()
    revealed=bg.copy()
    for c in range(0, 3):
        hidden[y1:y2, x1:x2, c] = (malpha_s * mask[:, :, c] + malpha_l * hidden[y1:y2, x1:x2, c])
        revealed[y1:y2, x1:x2, c] = (alpha_s * img[:, :, c] + alpha_l * revealed[y1:y2, x1:x2, c])
    
    
    cv2.imwrite('hidden/'+str(pokedex_number)+'.png', hidden)
    cv2.imwrite('revealed/'+str(pokedex_number)+'.png', revealed)
```
